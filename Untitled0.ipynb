```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import random
```

```python
IMG_SIZE = 128

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset/train",
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=32
)

test_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset/test",
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=32
)

class_names = train_dataset.class_names
print(class_names)
```

```python
normalization_layer = layers.Rescaling(1./255)

train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
test_dataset = test_dataset.map(lambda x, y: (normalization_layer(x), y))
```

```python
def build_cnn(filters, lr, dropout):

    model = models.Sequential([
        layers.Conv2D(filters, (3,3), activation='relu',
                      input_shape=(IMG_SIZE, IMG_SIZE, 3)),
        layers.MaxPooling2D(),

        layers.Conv2D(filters*2, (3,3), activation='relu'),
        layers.MaxPooling2D(),

        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(dropout),

        layers.Dense(len(class_names), activation='softmax')
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
def fitness(solution):

    filters = int(solution[0])
    lr = solution[1]
    dropout = solution[2]
    batch = int(solution[3])

    model = build_cnn(filters, lr, dropout)

    history = model.fit(
        train_dataset,
        validation_data=test_dataset,
        epochs=3,
        verbose=0
    )

    val_acc = history.history['val_accuracy'][-1]

    return val_acc
```

```python
class RCCO:

    def __init__(self, pop_size=4, iterations=3):

        self.pop_size = pop_size
        self.iterations = iterations

        self.bounds = [
            (16, 64),        # filters
            (0.0001, 0.01),  # learning rate
            (0.2, 0.5),      # dropout
            (16, 64)         # batch size
        ]

    def init_population(self):

        pop = []

        for _ in range(self.pop_size):
            individual = [
                random.uniform(*self.bounds[i])
                for i in range(len(self.bounds))
            ]
            pop.append(individual)

        return pop

    def update(self, crane, best):

        new = []

        for i in range(len(crane)):

            r = random.random()
            value = crane[i] + r * (best[i] - crane[i])

            low, high = self.bounds[i]
            value = np.clip(value, low, high)

            new.append(value)

        return new

    def optimize(self):

        population = self.init_population()

        best_sol = None
        best_fit = -1

        for it in range(self.iterations):

            print("Iteration", it+1)

            for ind in population:

                fit = fitness(ind)

                if fit > best_fit:
                    best_fit = fit
                    best_sol = ind

            population = [self.update(ind, best_sol) for ind in population]

            print("Best accuracy :", best_fit)

        return best_sol
```

```python
optimizer = RCCO()

best_solution = optimizer.optimize()

print("Meilleurs paramètres trouvés :", best_solution)
```

```python
filters = int(best_solution[0])
lr = best_solution[1]
dropout = best_solution[2]

final_model = build_cnn(filters, lr, dropout)

history = final_model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=10
)
```
